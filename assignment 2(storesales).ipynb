```python
2 ## check shape , data types and non-null counts
print(data.shape)  ## shape
print(data.info()) ## data types
print(data.isnull().sum())  ### null couts
```

```python
## print column names and summary statics for all numeric columns
print(data.columns.tolist()) ## all columns names
print(data.select_dtypes(include=['number'])) ### all numeric columns 
print(data.describe()) ## summary of all the numeric columns
```

```python
### find the total and average sales for each month (jan-april)
a= data[['Jan','Feb','Mar','Apr']].agg(['mean','sum'])
a
```

```python
### calculate total yearly sales for all stored combines
total_sales = data.loc[:,'Jan':'Dec'].sum().sum()
total_sales
```

```python
## find total sales for each city in each month using groupby()
city_monthly_sales = data.groupby('city')[['Jan','Feb','Mar','Apr','May','Jun','July','Aug','Sep','Oct','Nov','Dec']].sum()
city_monthly_sales
```

```python
## find average monthly sales per city
city_sales= data.groupby('city')[['Jan','Feb','Mar','Apr','May','Jun','July','Aug','Sep','Oct','Nov','Dec']].mean()
city_sales
```

```python
### the city with the highest average sales
avg = city_sales.mean(axis=1).sort_values(ascending=False)
avg
```

```python
## add the new column total_sales = jan+feb+mar+apr
data['Total_sales'] = data[['Jan','Feb','Mar','Apr']].sum(axis=1)
data
```

```python
##find the top 3 performing cities based on total sales
city = data.groupby('city').sum(numeric_only=True).sum(axis=1)
cities = city.sort_values(ascending=False).head(3)
print(cities)
```
