```python
# LOADING THE UBER.CSV DATA SET
import pandas as pd  
uber_data=pd.read_csv("uber.csv")
uber_data.head()
```

```python
##  RENAME ALL THE COLUMNS IN UPPERCASE
#  RENAME ALL THE COLUMNS IN UPPERCASE
uber_data = uber_data.rename(columns={'START_DATE*': 'START','END_DATE*': 'END','CATEGORY*': 'CATEGORY','START*': 'STA','STOP*': 'STP','MILES*': 'MILES','PURPOSE*': 'PURPOSE'})
uber_data
```

```python
### ASSIGNMENT 2
# 1. DISPLAY RIDES WHERE CATEGORY == 'business'
rides=uber_data[uber_data['CATEGORY']=='Business']
rides
```

```python
## 2. SHOW THE TOP 5 RIDES WITH THE LONGEST DISTANCE(miles)
a = uber_data.sort_values(by='MILES', ascending=False).head(6) 
a
```

```python
# 3.REPLACE ALL THE MISSING PURPOSE VALUES WITH "NOT SPECIFIED"
uber_data.loc[uber_data['PURPOSE'] == 'Not Specified']
uber_data
```

```python
# 4. CREATE A NEW COLUMN TRIP_DURATION USING END_DATA-START_DATA
uber_data['START'] = pd.to_datetime(uber_data['START'], errors='coerce')
uber_data['END'] = pd.to_datetime(uber_data['END'], errors='coerce')
uber_data['TRIP_DURATION'] = uber_data['END'] - uber_data['START']
uber_data
```

```python
# 5. sort trips by distance in acsending order
a = uber_data.sort_values(by='MILES', ascending=True) 
a
```
